## Exploring df_start_end

```python
df_start_end = pd.read_csv('../assets/df_start_end.csv', index_col=0)
# example of read_csv on google drive
# df_diff = pd.read_csv('/content/drive/MyDrive/Empresas/ZRP/Desafio Técnico/zrp_case-main/assets/df_diff.csv')
# model = keras.models.load_model('/content/drive/MyDrive/Empresas/ZRP/Desafio Técnico/zrp_case-main/assets/model')

df_start_end
```

```python
df_min_max = pd.DataFrame(index=['min', 'max'])
for name, column in df_start_end.iteritems():
    min_max = pd.Series([column.min(), column.max()], index=['min', 'max'])
    df_min_max[name] = min_max

df_min_max
```

**Notes**:
- The table above shows that each read have the same range of values.

## Exploring df_diff

```python
df_diff = pd.read_csv('../assets/df_diff.csv', index_col=0)
df_diff
```

```python
# the following columns will be dropped due to the fact that they are not relevant to the analysis due to the constant nature through the rows of dataset
df_diff.drop(columns=['start_timestamp', 'end_timestamp', 'diff_timestemp'], inplace=True)

df_diff.describe()
```

```python
display(df_diff.corr())
scatter_matrix(df_diff, figsize=(12, 12))
plt.show()
```
